# **Python** **Income** **Tax** **Calculator**

```python
def Income_Tax_Calculator(income):
  tax = 0

  if income <= 250000:
    tax = 0


  elif income <= 500000:
    tax = (income - 250000) * 0.05


  elif income <= 750000:
    tax = (250000*0.07)+(income-500000)*0.10


  elif income <= 1000000 :
     tax = (250000*0.05) +(income - 500000)*0.20


  else:
      tax = (250000 * 0.05) + (500000 * 0.02) + (income - 1000000) * 0.03

  return tax


salary = float(input("Enter your annual income : "))


result = Income_Tax_Calculator(salary)


print("Your income tax is : ", result)
```
